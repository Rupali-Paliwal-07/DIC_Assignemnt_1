```python
%sql



CREATE CATALOG IF NOT EXISTS main;

-- Verify catalog created
SHOW CATALOGS;
-- Should show: main, system (and any others)



CREATE SCHEMA IF NOT EXISTS main.cve_assignment


-- Verify schema
SHOW SCHEMAS IN main;
-- Should show: cve_assignment, default



CREATE VOLUME IF NOT EXISTS main.cve_assignment.cve_data


-- Verify volume
DESCRIBE VOLUME main.cve_assignment.cve_data;
LIST '/Volumes/main/cve_assignment/cve_data/';



CREATE DATABASE IF NOT EXISTS main.cve_bronze


CREATE DATABASE IF NOT EXISTS main.cve_silver


CREATE DATABASE IF NOT EXISTS main.cve_gold



SHOW DATABASES IN main;



SELECT 'Catalogs' as resource, COUNT(*) as count FROM system.information_schema.catalogs WHERE catalog_name = 'main'
UNION ALL
SELECT 'Schemas', COUNT(*) FROM system.information_schema.schemata WHERE catalog_name = 'main'
UNION ALL
SELECT 'Databases in main', COUNT(*) FROM system.information_schema.schemata 
  WHERE catalog_name = 'main' AND schema_name LIKE 'cve_%';

```
